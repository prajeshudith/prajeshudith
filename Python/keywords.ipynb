```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from datetime import date
from googletrans import Translator
```

```python
news_articles = []
keyword = input('Enter the word:')
luces = "https://lucesdelsiglo.com/?s="+keyword
r = requests.get(luces)

soup = BeautifulSoup(r.content, 'html.parser')
```

```python
r
```

```python
news = soup.findAll('div',{'class':'item-details'})
```

```python
link = []
for i in range(len(news)):
      url = news[i].find('a')['href']
      link.append(url)
      
```

```python
headlines = []
for i in range(len(news)):
     title = news[i].find('a').text
     headlines.append(title)
```

```python
def bert_extractor(x):
  kb_model = KeyBERT()
  words = kb_model.extract_keywords(x,keyphrase_ngram_range=(1,1),stop_words=None)
  phrases = []
  phrases.append(words)
  return phrases
```

```python
col_list = []

# luces contains all the sentences from all the news
for i in range(len(luces)):
    
    sent = luces.loc[i,'Translated_Abstract']
    col_list.append(list(bert_extractor(sent)))
```

```python
col_list
```

```python
bert_words  = pd.DataFrame({'Bert_Keyword':col_list})
bert_words
```

```python
import itertools

regular_list = bert_words['Bert_Keyword'][0]
flat_list = list(itertools.chain(*regular_list))
flat_list
fl_lst = []
for lst in bert_words['Bert_Keyword']:
  flat_list = list(itertools.chain(*lst))
  bert = list(itertools.chain(*flat_list))
  fl_lst.append(bert)
fl_lst
```
