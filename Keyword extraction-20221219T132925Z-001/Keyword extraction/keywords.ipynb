```python
def bert_extractor(x):
  kb_model = KeyBERT()
  words = kb_model.extract_keywords(x,keyphrase_ngram_range=(1,1),stop_words=None)
  phrases = []
  phrases.append(words)
  return phrases
```

```python
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    #Disctionary object created
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
    """
    print("Overall sentiment dictionary is : ", sentiment_dict)
    print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
    """
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        #print("Positive")
        return "Positive"
 
    elif sentiment_dict['compound'] <= - 0.05 :
        #print("Negative")
        return "Negative"
 
    else :
        #print("Neutral")
        return "Neutral"
```

```python
#Saves the final dataframe as a file and collects into a dictionary
cnt = 1
final_df = {}

for i in qw:
  j = 0
  search_words = "cancun fc"

  for k in range(len(i)):
    if not int(k) and j < 2:
       search_words = search_words + "OR" + i[k]
       j += 1

    if j == 2:
      break
    
  new_search = search_words + " -filter:retweets"
  date_since = "2022-01-12"
  date_until = "2022-01-19"

  tweets = tweepy.Cursor(api.search,
                q=search_words,
                lang="es",
                since=date_since,
                until=date_until).items()

  final_data = [[tweet.user.location, tweet.created_at, tweet.text, tweet.favorite_count, tweet.retweet_count] for tweet in tweets]

  tweet_text = pd.DataFrame(data=final_data, columns=["Location", "Time", "Tweet", "Favorite Count", "Retweet Count"])
  name_file = "tweets" + str(cnt) +".csv"
  tweet_text.to_csv(name_file)

  final_cnts = []

  for i in range(len(tweet_text)):
      final_cnts.append(sentiment_scores(tweet_text['Tweet'][i]))

  final_cnt_df = Counter(final_cnts)
  ans = ""

  if final_cnt_df['Positive'] >= final_cnt_df['Positive']:
    ans = "Positive"
  else:
    ans = "Negative" 

  final_df[cnt] = [name_file, ans]
```
