### Importing required libraries

```python
from datetime import date
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

### Connecting and extracting the content from the given url 

```python
keyword = "messi+ronaldo"
url = "https://www.bbc.co.uk/search?q=" + keyword
r1 = requests.get(url)
coverpage = r1.content
```

```python
soup1 = BeautifulSoup(coverpage, 'html5lib')
```

```python
coverpage_news = soup1.find_all('div', {'class': 'ssrcss-1cbga70-Stack e1y4nx260'})
```

### Fixing Keyword and initialising the required lists 

```python
title = []
links = []
dates = []
description = []
```

### Collecting the title, date and links from the filtered articles 

```python
j = 0
for i in range(1,len(coverpage_news)):
    print(i)
    if coverpage_news[i].find('a')['href'][0:2] == "/s":
        url_ = "https://www.bbc.com/" + coverpage_news[i].find('a')['href']
    else:
        url_ = coverpage_news[i].find('a')['href']
    
    r = requests.get(url_)
    cover_page = r.content
    soup1 = BeautifulSoup(cover_page, 'html5lib')

    if "https://www.bbc.co.uk/sounds/play" in url_:
        coverpagenews1 = soup1.find_all('div', {'class': 'gel-pica gs-u-display-inline@s sc-c-episode__metadata__data'})
        temp = coverpagenews1[0].get_text().split(":")
        temp1 = temp[-1]
        temp1 = temp1.lstrip()
        dates.append(temp1)
    else:
        dates.append(new[j])
        j += 1
        
    description.append(coverpage_news[i].find('p', {'class' : 'ssrcss-1q0x1qg-Paragraph eq5iqo00'}).get_text())
    title.append(coverpage_news[i].find('span').get_text())
    links.append(url_)
    
```

```python
dict = {'Title': title, 'Link': links, 'Date': dates, 'Extracted date': current_dates, 'Description': description} 
df = pd.DataFrame(dict) 
df.to_csv('data_list_bbc.csv')
```
